# ECC from pointcloud using alpha complexes

```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
from pyEulerCurves import ECC_from_pointcloud, plot_euler_curve
```

## Sample points from the unit circle

```python
point_cloud = []

NUMBER_OF_POINTS = 100

random.seed(42)
for i in range(NUMBER_OF_POINTS):
    angle = random.uniform(0, 2 * np.pi)
    point_cloud.append([np.cos(angle), np.sin(angle)])

point_cloud = np.array(point_cloud)
```

```python
plt.figure()
plt.axis("scaled")
plt.xlim((-1.2, 1.2))
plt.ylim((-1.2, 1.2))
plt.scatter(point_cloud[:, 0], point_cloud[:, 1])
plt.show()
```

## Alpha ECC

```python
trans = ECC_from_pointcloud(complex_type="alpha")
```

```python
%%time
alpha_ecc = trans.fit_transform(point_cloud)
```

```python
# note how the EC is constant after epsilon=0.3
fig, axs = plt.subplots(1)

im0 = plot_euler_curve(alpha_ecc, axs, with_lines=True)
```
